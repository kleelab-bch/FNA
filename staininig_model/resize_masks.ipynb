## Generate resized masks for object detection

```python
import tensorflow as tf
import tensorflow_datasets as tfds
import cv2
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "1"
import time
import matplotlib.pyplot as plt
from IPython.display import clear_output

tfds.disable_progress_bar()
AUTOTUNE = tf.data.experimental.AUTOTUNE
```

```python
train_root_mask_path = '../../TfResearch/research/object_detection/dataset_tools/assets/masks_train/'
valid_root_mask_path = '../../TfResearch/research/object_detection/dataset_tools/assets/masks_valid/'
test_root_mask_path = '../../TfResearch/research/object_detection/dataset_tools/assets/masks_test/'

def load_image(image_file):
    image_name = tf.strings.regex_replace(image_file, train_root_mask_path, '')
    image_name = tf.strings.regex_replace(image_name, valid_root_mask_path, '')
    image_name = tf.strings.regex_replace(image_name, test_root_mask_path, '')
    
    image = tf.io.read_file(image_file)
    image = tf.image.decode_png(image, channels=3)
    image = tf.cast(image, tf.float32)

    return (image, image_name)


def resize(input_image, height, width):
  input_image = tf.image.resize(input_image, [height, width],
                                method=tf.image.ResizeMethod.NEAREST_NEIGHBOR)

  return input_image
```

```python
root_image_path = '../../TfResearch/research/object_detection/dataset_tools/assets'
masks_train_dataset_names = tf.data.Dataset.list_files(root_image_path+'/masks_train/*.png')
masks_valid_dataset_names = tf.data.Dataset.list_files(root_image_path+'/masks_valid/*.png')
masks_test_dataset_names = tf.data.Dataset.list_files(root_image_path+'/masks_test/*.png')

train_dataset = masks_train_dataset_names.map(load_image, num_parallel_calls=tf.data.experimental.AUTOTUNE).cache().batch(1)
valid_dataset = masks_valid_dataset_names.map(load_image, num_parallel_calls=tf.data.experimental.AUTOTUNE).cache().batch(1)
test_dataset = masks_test_dataset_names.map(load_image, num_parallel_calls=tf.data.experimental.AUTOTUNE).cache().batch(1)
```

```python
train_save_path = 'resized_masks_train/'
valid_save_path = 'resized_masks_valid/'
test_save_path = 'resized_masks_test/'

def save_image(save_path, inp):
    input_image = inp[0]
    image_name = inp[1].numpy()[0].decode("utf-8")
    input_image = resize(input_image[0], 1792, 2560)
    input_image = input_image.numpy()
    cv2.imwrite(save_path + image_name, cv2.cvtColor(input_image, cv2.COLOR_RGB2BGR))
```

```python
data_num = train_dataset.cardinality().numpy()
print(data_num)
for index, inp in enumerate(train_dataset.take(data_num)):
    print(index)
    save_image(train_save_path, inp)
```

```python
data_num = valid_dataset.cardinality().numpy()
print(data_num)
for index, inp in enumerate(valid_dataset.take(data_num)):
    print(index)
    save_image(valid_save_path, inp)
```

```python
data_num = test_dataset.cardinality().numpy()
print(data_num)
for index, inp in enumerate(test_dataset.take(data_num)):
    print(index)
    save_image(test_save_path, inp)
```
